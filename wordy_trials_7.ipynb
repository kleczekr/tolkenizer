```
import spacy
from spacy.matcher import Matcher
from tabulate import tabulate
from itertools import zip_longest
```

```
nlp = spacy.load("en_core_web_sm")
```

```
# This cell is meant to accommodate the Google Colab way of dealing with reading 
# files from Google Drive; feel free to ignore it if you are running the notebook
# on your local machine
from google.colab import drive
drive.mount('/content/drive')
```

```
with  open('/content/drive/My Drive/book_txt/tbl_gbr.txt', 'r') as f:
  jkr = f.read()
```

```
nlp.max_length = 1709789
```

```
jkr_doc = nlp(jkr)
```

```
# the match patterns
strike_pattern = [{'LOWER': 'strike'}, {'POS': 'VERB'}]
rob_pattern = [{'LOWER': 'robin'}, {'POS': 'VERB'}]
```

```
matcher = Matcher(nlp.vocab)
matcher.add('STRIKE', None, strike_pattern)
matcher.add('ROBIN', None, rob_pattern)
```

```
robin_dict = dict()
strike_dict = dict()
```

```
for match_id, start, end in matcher(jkr_doc):
  if jkr_doc.vocab.strings[match_id] == 'STRIKE':
    if nlp(jkr_doc[start:end].text)[1].lemma_ not in strike_dict.keys():
      strike_dict[nlp(jkr_doc[start:end].text)[1].lemma_] = 1
    else:
      strike_dict[nlp(jkr_doc[start:end].text)[1].lemma_] += 1
  else:
    if nlp(jkr_doc[start:end].text)[1].lemma_ not in robin_dict.keys():
      robin_dict[nlp(jkr_doc[start:end].text)[1].lemma_] = 1
    else:
      robin_dict[nlp(jkr_doc[start:end].text)[1].lemma_] += 1
```

```
sorted_robin = [(k, v) for k, v in sorted(robin_dict.items(), key=lambda x: x[1], reverse=True)]
sorted_strike = [(k, v) for k, v in sorted(strike_dict.items(), key=lambda x: x[1], reverse=True)]
```

```
print(sorted_robin)
```

```
for item in sorted_robin:
  print(item)
```

```
# convert sets in the list to lists
sorted_robin = [list(item) for item in sorted_robin]
sorted_strike = [list(item) for item in sorted_strike]
```

```
strike_robin_counted = zip_longest(sorted_strike, sorted_robin, fillvalue=' ')
```

```
strike_robin_counted = [list(stuff) for stuff in strike_robin_counted]
```

```
for item in strike_robin_counted:
  print(item)
```

```
headers_ = ['Strike', 'Robin']
print(tabulate(strike_robin_counted, headers_))
```

```
with  open('/content/drive/My Drive/book_txt/killing_floor.txt', 'r') as f:
  meh = f.read()
```

```
meh_doc = nlp(meh)
```

```
meh_pattern = [{'IS_TITLE': True, 'POS': 'PROPN'}, {'POS': 'VERB'}]
```

```
matcher = Matcher(nlp.vocab)
matcher.add('PATTERN', None, meh_pattern)
```

```
meh_dict = dict()
```

```
for match_id, start, end in matcher(meh_doc):
  if nlp(meh_doc[start:end].text)[1].lemma_ not in meh_dict.keys():
    meh_dict[nlp(meh_doc[start:end].text)[1].lemma_] = 1
  else:
    meh_dict[nlp(meh_doc[start:end].text)[1].lemma_] += 1
```

```
sorted_meh = [[k, v] for k, v in sorted(meh_dict.items(), key=lambda x: x[1], reverse=True)]
```

```
headers = ['word', 'count']
print(tabulate(sorted_meh, headers))
```
