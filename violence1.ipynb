```
pd.set_option('display.max_rows', 500)
```

```
# display the mean number of casualties for a single
# incident in each country
df.groupby('country_txt').nkill.agg(['mean', 'sum'])
```

```
# display the total number of casualties in all the
# incidents mentioned for a given country
df.groupby('country_txt').nkill.sum()
```

```
# display the total number of casualties in all the
# incidents mentioned for a given region
df.groupby('region_txt').nkill.sum()
```

```
%matplotlib inline
```

```
df.groupby('region_txt')['nkill', 'nwound'].sum().plot(kind='bar')
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot']) # optional: for ggplot-like style


df.groupby('region_txt')['nkill', 'nwound'].sum().plot(kind='bar', figsize=(20, 10))

plt.title('Total number of killed and wounded per region')
plt.ylabel('Total number of victims')
plt.xlabel('Region')
# to tilt the bottom labels:
plt.xticks(rotation=30, ha='right')
plt.show()
```
