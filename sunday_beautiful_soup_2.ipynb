```
import pandas as pd
```

```
link_sa = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/southasia'
link_sl = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/srilanka'
link_bd = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/bangladesh'
link_bh = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/bhutan'
link_in = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/india'
link_np = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/nepal'
link_ps = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/pakistan'
```

```
sl_df = pd.read_html(link_sl)[1]
```

```
sl_df
```

```
df_sa = pd.read_html(link_sa)[1]
df_sl = pd.read_html(link_sl)[1]
df_bd = pd.read_html(link_bd)[1]
df_bh = pd.read_html(link_bh)[1]
df_in = pd.read_html(link_in)[1]
df_np = pd.read_html(link_np)[1]
df_ps = pd.read_html(link_ps)[1]
```

```
df_sa['country'] = 'South Asia total'
df_sl['country'] = 'Sri Lanka'
df_bd['country'] = 'Bangladesh'
df_bh['country'] = 'Bhutan'
df_in['country'] = 'India'
df_np['country'] = 'Nepal'
df_ps['country'] = 'Pakistan'
```

```
df = pd.concat([df_sa, df_sl, df_bd, df_bh, df_in, df_np, df_ps])
```

```
df = df.reset_index(drop=True)
```

```
df.head(40)
```

```
df = df[df.Year!='Total**']
```

```
df.columns = ['year', 'incidents', 'civilian_casualties', 'sf_casualties', 'terrorist_casualties', 'unspecified_casualties', 'total_casualties', 'country']
```

```
df.year = df.year.map(lambda x: x.rstrip('*'))
```

```
df = df.astype({'year': int, 'incidents': int, 'civilian_casualties': int, 'sf_casualties': int, 'terrorist_casualties': int, 'unspecified_casualties': int, 'total_casualties': int})
```
