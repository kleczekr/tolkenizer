```
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```
link_sa = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/southasia'
link_sl = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/srilanka'
link_bd = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/bangladesh'
link_bh = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/bhutan'
link_in = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/india'
link_np = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/nepal'
link_ps = 'https://www.satp.org/datasheet-terrorist-attack/fatalities/pakistan'
```

```
def pipeline_table(link):
  # read the HTML with urllib.request
  page = urllib.request.urlopen(link)
  # now, parse the HTML with soup
  soup = BeautifulSoup(page, 'html.parser')
  # find the table
  table = soup.find('table', id='ctl00_ContentPlaceHolder1_gvFatalities')
  table_headers = table.find_all('th')
  table_rows = table.find_all('tr')
  # find headers
  h = []
  for header in table_headers:
    header = header.text
    h.append(header)
  # save the rows
  l = []
  for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td]
    l.append(row)
  return pd.DataFrame(l[1:], columns=h)
```

```
# links = [link_sa, link_sl, link_bd, link_bh, link_in, link_np, link_ps]
# dataframe_names = [df_sa, df_sl, df_bd, df_bh, df_in, df_np, df_ps]
```

```
df_sa = pipeline_table(link_sa)
df_sl = pipeline_table(link_sl)
df_bd = pipeline_table(link_bd)
df_bh = pipeline_table(link_bh)
df_in = pipeline_table(link_in)
df_np = pipeline_table(link_np)
df_ps = pipeline_table(link_ps)
```
