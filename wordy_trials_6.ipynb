```
import spacy
from spacy.matcher import Matcher
from itertools import zip_longest
from tabulate import tabulate
```

```
nlp = spacy.load("en_core_web_sm")
```

```
# This cell is meant to accommodate the Google Colab way of dealing with reading 
# files from Google Drive; feel free to ignore it if you are running the notebook
# on your local machine
from google.colab import drive
drive.mount('/content/drive')
```

```
with  open('/content/drive/My Drive/book_txt/tbl_gbr.txt', 'r') as f:
  jkr = f.read()
```

```
nlp.max_length = 1709789
```

```
jkr_doc = nlp(jkr)
```

```
# the match patterns
strike_pattern = [{'LOWER': 'strike'}, {'POS': 'VERB'}]
rob_pattern = [{'LOWER': 'robin'}, {'POS': 'VERB'}]
```

```
matcher = Matcher(nlp.vocab)
matcher.add('STRIKE', None, strike_pattern)
matcher.add('ROBIN', None, rob_pattern)
```

```
for match_id, start, end in matcher(jkr_doc):
  print(jkr_doc.vocab.strings[match_id], jkr_doc[start:end].text)
```

```
for match_id, start, end in matcher(jkr_doc):
  print(jkr_doc.vocab.strings[match_id], nlp(jkr_doc[start:end].text)[1].lemma_)
```

```
strike_set = set()
robin_set = set()
```

```
for match_id, start, end in matcher(jkr_doc):
  if jkr_doc.vocab.strings[match_id] == 'STRIKE':
    strike_set.add(nlp(jkr_doc[start:end].text)[1].lemma_)
  else:
    robin_set.add(nlp(jkr_doc[start:end].text)[1].lemma_)
```

```
print(f"List of Strike's entries counts {len(strike_set)} verbs.\n"
f"List of Robin's entries counts {len(robin_set)} verbs.")
```

```
strike_list = list(strike_set)
robin_list = list(robin_set)
# since the lists are of unequal length, import and use the zip_longest
# method instead of zip
strike_robin_list = zip_longest(sorted(strike_list), sorted(robin_list), fillvalue=' ')
# the zip function returns list of sets---convert them to strings
strike_robin_list = [list(item) for item in strike_robin_list]
```

```
headers_ = ['Strike', 'Robin']
print(tabulate(strike_robin_list, headers_))
```
