```
%matplotlib inline 

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```
# This cell is meant to accommodate the Google Colab way of dealing with reading 
# files from Google Drive; feel free to ignore it if you are running the notebook
# on your local machine
from google.colab import drive
drive.mount('/content/drive')
```

```
df = pd.read_excel('/content/drive/My Drive/ethnic_violence/globalterrorismdb_0919dist.xlsx')
```

```
gb_region = df.groupby(['iyear', 'region_txt']).nkill.sum()
gb_region[160:200]
```

```
gb_region.plot(kind='area',
                  stacked=False,
                  figsize=(20,10))
plt.title('Number of casualties per year per region')
plt.ylabel('total casualties')
plt.xlabel('year')
plt.show()
```

```
# Let's convert the gb_region series back to a datafame,
# with the second index
df_gb_region = gb_region.unstack(level=1)
```

```
df_gb_region.head(10)
```

```
df_gb_region.plot(kind='area',
                  stacked=False,
                  figsize=(20,10))
plt.title('Number of casualties per year per region')
plt.ylabel('total casualties')
plt.xlabel('year')
plt.show()
```

```
df_gb_region.plot(kind='area', 
                  alpha=0.75,
                  stacked=True,
                  figsize=(20,10))
plt.title('Number of casualties per year per region')
plt.ylabel('total casualties')
plt.xlabel('year')
plt.show()
```
