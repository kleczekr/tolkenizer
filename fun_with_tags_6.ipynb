```
from bs4 import BeautifulSoup
import urllib.request
```

```
link = 'http://gretil.sub.uni-goettingen.de/gretil.html'
# read the HTML with urllib.request
page = urllib.request.urlopen(link)
# now, parse the HTML with soup
soup = BeautifulSoup(page, 'html.parser')
```

```
# let's take a look at the website
print(soup.prettify())
```

```
for divider in soup.find_all('div'):
  if divider.get('id') == 'text-GrhyaS':
    for child in divider.ol.children:
      print(child)
```

```
for divider in soup.find_all('div'):
  if divider.get('id') == 'text-GrhyaS':
    for string in divider.ol.stripped_strings:
      print(string)
```

```
for divider in soup.find_all('div'):
  if divider.get('id') == 'text-GrhyaS':
    for string in divider.ol.find_all('li'):
      print(string)
```

```
for divider in soup.find_all('div'):
  if divider.get('id') == 'text-GrhyaS':
    for string in divider.ol.findChildren('li', recursive = False):
      print(string)
```
