```
df = pd.read_excel('/content/drive/My Drive/ethnic_violence/globalterrorismdb_0919dist.xlsx')
```

```
df.boxplot(by ='attacktype1_txt', column =['nkill'], grid = True)
```

```
import seaborn as sns
sns.set_style("whitegrid") 
  
sns.boxplot(x = 'attacktype1_txt', y = 'nkill', data = df) 
```

```
viz = sns.catplot("attacktype1_txt", "nkill",
                   data=df, kind="box",
                   height=10, palette="muted", legend=False)
viz.fig.get_axes()[0].set_yscale('log')
viz.set_xticklabels(rotation=45)
```

```
viz = sns.catplot("attacktype1_txt", "nkill",
                   data=df, kind="bar",
                   height=10, palette="muted", legend=False)
viz.set_xticklabels(rotation=45)
```
