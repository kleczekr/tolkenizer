```
import spacy
from spacy.matcher import Matcher
from tabulate import tabulate
from itertools import zip_longest
```

```
# This cell is meant to accommodate the Google Colab way of dealing with reading 
# files from Google Drive; feel free to ignore it if you are running the notebook
# on your local machine
from google.colab import drive
drive.mount('/content/drive')
```

```
nlp = spacy.load("en_core_web_sm")
```

```
with  open('/content/drive/My Drive/book_txt/hp/hp1.txt', 'r') as f:
  hp1 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp2.txt', 'r') as f:
  hp2 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp3.txt', 'r') as f:
  hp3 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp4.txt', 'r') as f:
  hp4 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp5.txt', 'r') as f:
  hp5 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp6.txt', 'r') as f:
  hp6 = f.read()
with  open('/content/drive/My Drive/book_txt/hp/hp7.txt', 'r') as f:
  hp7 = f.read()
```

```
nlp.max_length = 1493158
```

```
hp1_doc = nlp(hp1)
hp2_doc = nlp(hp2)
hp3_doc = nlp(hp3)
hp4_doc = nlp(hp4)
hp5_doc = nlp(hp5)
hp6_doc = nlp(hp6)
hp7_doc = nlp(hp7)
```

```
# the match patterns
harry_pattern = [{'LOWER': 'harry'}, {'POS': 'VERB'}]
ron_pattern = [{'LOWER': 'ron'}, {'POS': 'VERB'}]
hermione_pattern = [{'LOWER': 'hermione'}, {'POS': 'VERB'}]
```

```
matcher = Matcher(nlp.vocab)
matcher.add('HARRY', None, harry_pattern)
matcher.add('RON', None, ron_pattern)
matcher.add('HERMIONE', None, hermione_pattern)
```

```
counter = 1
```

```
def verbalizer_pipeline(doc_name):
  harry_dict = {}
  ron_dict = {}
  hermione_dict = {}

  for match_id, start, end in matcher(doc_name):
    if doc_name.vocab.strings[match_id] == 'HARRY':
      if nlp(doc_name[start:end].text)[1].lemma_ not in harry_dict.keys():
        harry_dict[nlp(doc_name[start:end].text)[1].lemma_] = 1
      else:
        harry_dict[nlp(doc_name[start:end].text)[1].lemma_] += 1
    elif doc_name.vocab.strings[match_id] == 'RON':
      if nlp(doc_name[start:end].text)[1].lemma_ not in ron_dict.keys():
        ron_dict[nlp(doc_name[start:end].text)[1].lemma_] = 1
      else:
        ron_dict[nlp(doc_name[start:end].text)[1].lemma_] += 1
    else:
      if nlp(doc_name[start:end].text)[1].lemma_ not in hermione_dict.keys():
        hermione_dict[nlp(doc_name[start:end].text)[1].lemma_] = 1
      else:
        hermione_dict[nlp(doc_name[start:end].text)[1].lemma_] += 1

  sorted_harry = [[k, v] for k, v in sorted(harry_dict.items(), key=lambda x: x[1], reverse=True)]
  sorted_ron = [[k, v] for k, v in sorted(ron_dict.items(), key=lambda x: x[1], reverse=True)]
  sorted_hermione = [[k, v] for k, v in sorted(hermione_dict.items(), key=lambda x: x[1], reverse=True)]

  hrh_counted = zip_longest(sorted_harry, sorted_ron, sorted_hermione, fillvalue=' ')
  hrh_counted = [list(stuff) for stuff in hrh_counted]

  headers_ = ['Harry', 'Ron', 'Hermione']
  print('\nVerb counts for main characters in the volume {} of Harry Potter\n'.format(counter))
  print(tabulate(hrh_counted, headers_))
```

```
list_of_harrys = [hp1_doc, hp2_doc, hp3_doc, hp4_doc, hp5_doc, hp6_doc, hp7_doc]
```

```
for harry in list_of_harrys:
  verbalizer_pipeline(harry)
  counter += 1
```
